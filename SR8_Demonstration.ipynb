```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

testimg=cv2.imread("car.jpg")
```

```python
testimg = cv2.cvtColor(testimg, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(testimg)
```

```python
testimg_hr = cv2.resize(testimg, (384, 384))
testimg_lr= cv2.resize(testimg, (96, 96))
```

```python
plt.imshow(testimg_hr)
```

```python
plt.imshow(testimg_lr)
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras import layers, Model
from sklearn.model_selection import train_test_split
from keras import Model
from keras.layers import Conv2D, PReLU,BatchNormalization, Flatten, Add, Lambda
from keras.layers import UpSampling2D, LeakyReLU, Dense, Input, add, Concatenate
from tqdm import tqdm
import tensorflow
from keras.applications import VGG19
```

```python
from keras.models import load_model
from numpy.random import randint
generator = load_model('gen_model500.h5', compile=False)
```

```python
import matplotlib.pyplot as plt
```

```python
sreeni_lr = testimg_lr / 255.
sreeni_hr = testimg_hr / 255.

sreeni_lr = np.expand_dims(sreeni_lr, axis=0)
sreeni_hr = np.expand_dims(sreeni_hr, axis=0)

generated_sreeni_hr = generator.predict(sreeni_lr)


# plot all three images

fig= plt.figure(figsize=(16, 8))
plt.subplot(231)
plt.title('LR Image')
plt.imshow(sreeni_lr[0,:,:,:])
img= plt.subplot(232)
plt.title('Superresolution')
plt.imshow(generated_sreeni_hr[0,:,:,:])
# extent = img.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
# #fig.savefig('ax2_figure.jpg', bbox_inches=extent)
# fig.savefig('ax2_figure_expanded3.jpg', bbox_inches=extent.expanded(1.1, 1.2))
plt.subplot(233)
plt.title('Orig. HR image')
plt.imshow(sreeni_hr[0,:,:,:])

plt.show()
```
